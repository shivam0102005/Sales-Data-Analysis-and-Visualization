```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Sales.csv', parse_dates=['Date'])
df.head()
```

```python
# Summery Statistics
print("Total Sales:",df['Total_Sales'].sum())
```

```python
print("Total Profit:",df['Profit'].sum())
```

```python
# Most Profitable product=> which gives most profit

print("Most profit Product",df['Profit'].max()) # for simple vision 
print("ðŸ”¸ Most Profitable Product:\n", df.loc[df['Profit'].idxmax()])# By using this we can see all detail of the product
```

```python
#Daily total sales and profit
Daily_sales=df.groupby('Date')["Total_Sales"].sum().reset_index()
daily_profits=df.groupby('Date')["Profit"].sum().reset_index()
```

```python
print(Daily_sales)
print("\n")
print(daily_profits)
```

```python
#monthly sales and profit

df['Month']=df["Date"].dt.to_period("M")
monthly_sales=df.groupby("Month")["Total_Sales"].sum().reset_index()
monthly_sales
```

```python
monthly_profit=df.groupby("Month")["Profit"].sum().reset_index()
monthly_profit
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(monthly_sales['Month'].astype(str), monthly_sales['Total_Sales'], marker='o', color='green')
plt.title('Monthly Total Sales')
plt.xlabel('Month')
plt.ylabel('Sales (â‚¹)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Category-wise comparison
import pandas as pd
import matplotlib.pyplot as plt

# Group by Category and calculate total Sales and Profit
category_summary = df.groupby('Category')[['Total_Sales', 'Profit']].sum().reset_index()

print("Category-wise Summary:\n", category_summary)

#  Plot side-by-side bar chart
plt.figure(figsize=(10,6))
bar_width = 0.35
x = range(len(category_summary['Category']))

# Plot sales
plt.bar(x, category_summary['Total_Sales'], width=bar_width, label='Total Sales')

# Plot profit 
plt.bar([i + bar_width for i in x], category_summary['Profit'], width=bar_width, label='Profit')

# X-axis labels and title
plt.xlabel('Category')
plt.ylabel('Amount ')
plt.title('Category-wise Sales vs Profit')
plt.xticks([i + bar_width/2 for i in x], category_summary['Category'])
plt.legend()
plt.tight_layout()

```

```python
#Pie chart for category wise sales and profits


total_sales=df.groupby("Category")["Total_Sales"].sum().reset_index()
total_sales
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Step 2: Group by Category and sum the sales
total_sales = df.groupby("Category")["Total_Sales"].sum().reset_index()

# Step 3: Plot pie chart
plt.figure(figsize=(8,8))
plt.pie(
    total_sales['Total_Sales'], 
    labels=total_sales['Category'], 
    autopct='%1.1f%%', 
    shadow=True
)

plt.title(' Category-wise Total Sales Distribution')
plt.axis('equal') 
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Step 2: Group by Category and sum the sales
total_profits = df.groupby("Category")["Profit"].sum().reset_index()

# Step 3: Plot pie chart
plt.figure(figsize=(8,8))
plt.pie(
    total_profits['Profit'], 
    labels=total_profits['Category'], 
    autopct='%1.1f%%', 
    shadow=True
)

plt.title(' Category-wise Total Profit Distribution')
plt.axis('equal') 
plt.tight_layout()
plt.show()
```

```python
#Daily trend in line graph


# Step 2: Group by Date and calculate total sales per day
# daily_sales = df.groupby('Date')['Total_Sales'].sum().reset_index()
daily_sales = df.groupby('Date')['Total_Sales'].sum()

# Step 3: Plot Line Graph
plt.figure(figsize=(10,6))

# plt.plot(daily_sales['Date'], daily_sales['Total_Sales'], marker='o', linestyle='-', color='blue')
plt.plot(daily_sales,marker='o',color='blue')

plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales (â‚¹)')
plt.grid(True)
plt.tight_layout()
```
